# Load Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score, silhouette_score
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import davies_bouldin_score, silhouette_score
```

```python
# Load Datasets
customers_df = pd.read_csv("Datasets/Customers.csv")
transactions_df = pd.read_csv("Datasets/Transactions.csv")
```

# Merge Data

```python
# Merge Data
merged_df = transactions_df.merge(customers_df, on="CustomerID")
```

# Feature Engineering

```python
# Feature Engineering: Aggregate Transaction Data per Customer
customer_features = merged_df.groupby("CustomerID").agg({
    "Price": "sum",  # Total money spent
    "ProductID": "nunique",  # Unique products purchased
    "Quantity": "sum"  # Total quantity bought
}).reset_index()
```

```python
# Normalize Data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(customer_features.iloc[:, 1:])
```

# Determine Optimal Number of Clusters (2 to 10)

```python
# Determine Optimal Number of Clusters (2 to 10)
db_scores = {}
silhouette_scores = {}

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(scaled_features)
    
    # Calculate Davies-Bouldin Index
    db_index = davies_bouldin_score(scaled_features, clusters)
    db_scores[k] = db_index
    
    # Calculate Silhouette Score
    silhouette = silhouette_score(scaled_features, clusters)
    silhouette_scores[k] = silhouette
```

```python
# Find Optimal Number of Clusters (Lowest DB Index)
optimal_k = min(db_scores, key=db_scores.get)
```

# Apply K-Means with Optimal K

```python
# Apply K-Means with Optimal K
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
customer_features["Cluster"] = kmeans.fit_predict(scaled_features)
```

```python
# Visualization 1 
plt.figure(figsize=(10, 5))
sns.scatterplot(x=customer_features["Price"], y=customer_features["Quantity"], hue=customer_features["Cluster"], palette="viridis")
plt.xlabel("Total Money Spent")
plt.ylabel("Total Quantity Bought")
plt.title("Customer Segmentation Clusters")
plt.legend(title="Cluster")
plt.show()
```

```python
# Visualization 2: Pairplot of Features by Cluster
sns.pairplot(customer_features, hue="Cluster", palette="husl")
plt.show()
```

# Results

```python
# Print Results
print(f"Optimal Number of Clusters: {optimal_k}")
print(f"Davies-Bouldin Index: {db_scores[optimal_k]:.4f}")
print(f"Silhouette Score: {silhouette_scores[optimal_k]:.4f}")
```

```python
# Save Clustering Results
customer_features.to_csv("Customer_Clusters.csv", index=False)
```
